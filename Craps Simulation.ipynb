## Here we will run our simulation 1,000 times and record the winning percentages.

```python
import random
import matplotlib.pyplot as plt


def roll():
    roll_one = random.randint(1, 6)
    roll_two = random.randint(1, 6)
    total_roll = roll_one + roll_two
    return total_roll


def game():
    pass_bets_wins = 0
    dont_pass_bets_wins = 0
    both_bets_lose = 0

    for i in range(0,1000):

        first_roll = roll()
        if first_roll == 7 or first_roll == 11:
            pass_bets_wins = pass_bets_wins + 1

        elif first_roll == 2 or first_roll == 3:
            dont_pass_bets_wins = dont_pass_bets_wins + 1
        elif first_roll == 12:
            both_bets_lose = both_bets_lose + 1
        else:
            while True:
                second_roll = roll()
                if second_roll == first_roll:
                    pass_bets_wins = pass_bets_wins + 1
                    break
                elif second_roll == 7:
                    dont_pass_bets_wins = dont_pass_bets_wins + 1
                    break
    print((pass_bets_wins/1000)*100)
    print((dont_pass_bets_wins/1000)*100)
    print((both_bets_lose/1000)*100)
    
    labels = 'Pass Bet Wins', 'Don\'t Pass Bet Wins', 'Both Bets Lose'
    sizes = [pass_bets_wins, dont_pass_bets_wins, both_bets_lose]
    explode = (0.1, 0.1, 0.1)

    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.2f%%', shadow=True, startangle=90)
    ax1.axis('equal')



game()
```
